```python
class Ensayo:

    def __init__(self,Fx,velocidad,H,L,fsampling=10):
        
        self.veloc = velocidad
        self.Fx = Fx
        self.H = H
        self.L = L
        self.rel_aspecto = L/H
        self.fsampling = fsampling
        self.F0 = 0
        self.time = 0
        self.rho = 1000
        self.Dsoporte = 30e-3
        self.Lsoporte = 50e-2
        
    def drag_teoria(self):
        CD = 1.3
        CDsoporte = 1
        FDx = CD*1/2*self.rho*self.veloc**2*self.L*self.H
        FDsoporte  = CDsoporte*1/2*self.rho*self.veloc**2*self.Dsoporte*self.Lsoporte
        
        self.Fx_teoria = FDx + FDsoporte
```

```python
dir_data = 'medidas7nov/'
files_txt = np.sort(glob.glob(dir_data+'*.txt'))

Ensayos = []
H = 15e-2
L = 30e-2
fsampling = 10
for i,filei in  enumerate(files_txt):
    A = pd.read_csv(filei,header=None)
    aux1 = np.nonzero(pd.Series(A.iloc[:,0]).str.contains('mV/V'))[0]
    Signal = np.array([float(ai.split('mV')[0]) for ai in A[0][aux1]])
    Fx = fit1(Signal)
    str_vel = filei.split('vel_')[-1][:-4]
    if filei.find('vel')==-1:
        velocidad = 0
    else:
        velocidad = float(str_vel)/100
    #if i==5:raise ValueError()
    Ensayos.append(Ensayo(Fx,velocidad,H,L,fsampling))
    Ensayos[i].drag_teoria()
```

```python
# valores previstos
```

```python
F0 = np.hstack((Ensayos[2].Fx,Ensayos[3].Fx)).mean()
for ensayoi  in Ensayos:
    ensayoi.F0 = F0
    ensayoi.time = np.arange(len(ensayoi.Fx))/ensayoi.fsampling
```

```python
# vista de los "ceros" medidos en el carro
fig,ax = plt.subplots()
for ens_i in Ensayos[:4]:
    ax.plot(ens_i.time,ens_i.Fx-ens_i.F0)
```

```python
fig1,ax1 = plt.subplots()

lin0, = ax1.plot(0,0,color='None',label='Veloc')
for i,ens_i in enumerate(Ensayos[4:]):
    lin1, = ax1.plot(ens_i.time,(ens_i.Fx-ens_i.F0)*1e-3,fillstyle='none',
                     label = f'{ens_i.veloc:.2f}',linestyle='',marker=mk[i])
    ax1.plot(ens_i.time,ens_i.time*0+ens_i.Fx_teoria,color=lin1.get_color())
ax1.legend()
ax1.set_xlabel('time [sec]')
ax1.set_ylabel('$F_x$ [kgf]');
```
