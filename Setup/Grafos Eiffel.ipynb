```python
def strg_to_float(lista,n):
    '''lista para convertir a valores flotantes, y n es el numero para cambiar las unidades pertinentes'''
    lista_num=[]
    for elem in lista:
        float_value = float(elem.replace(',', '.'))
        lista_num.append(float_value/n)
    return lista_num
```

```python
x_ar1_strg=[]
y_ar1_strg=[]



with open("Curva_Ar1.csv") as datos:
    lector=csv.reader(datos,delimiter=";")
    for row in lector:
        x_ar1_strg.append(row[0])
        y_ar1_strg.append(row[1])
    
x_ar1=strg_to_float(x_ar1_strg,1)
y_ar1=strg_to_float(y_ar1_strg,1)

x_ar2_strg=[]
y_ar2_strg=[]

with open("Curva_Ar2.csv") as datos:
    lector=csv.reader(datos,delimiter=";")
    for row in lector:
        x_ar2_strg.append(row[0])
        y_ar2_strg.append(row[1])

x_ar2=strg_to_float(x_ar2_strg,1)
y_ar2=strg_to_float(y_ar2_strg,1)

x_ar15_strg=[]
y_ar15_strg=[]

with open("Curva_Ar15.csv") as datos:
    lector=csv.reader(datos,delimiter=";")
    for row in lector:
        x_ar15_strg.append(row[0])
        y_ar15_strg.append(row[1])
        

x_ar15=strg_to_float(x_ar15_strg,1)
y_ar15=strg_to_float(y_ar15_strg,1)

print(x_ar15)


#Ploteo
x_start=0 
x_end=90
y_start=0 
y_end=1.5
plt.xlim ( x_start , x_end ) # Rango eje x
plt.ylim ( y_start , y_end ) # Rango eje y
plt.xlabel ("Angulo", fontsize =12) # Label eje x
plt.ylabel ("Cd", fontsize =12) # Label eje y

plt.title("Curvas de Drag")
plt.grid()
plt.plot(x_ar1,y_ar1)
plt.plot(x_ar15,y_ar15)
plt.plot(x_ar2,y_ar2)
#plt.show()
```
